```python
import numpy as np
import math as m
import matplotlib.pyplot as plt
import scipy
from scipy.linalg import lu_factor
x=[]
v=[]
Ek=[]
V=[]
C=[]
def pot(x):
    cos=-m.exp(-x**2)-1.2*m.exp(-(x-2)**2)
    return cos
x.append(2.8)
v.append(0)
t=0
dt=0.01
dx=0.01
for i in range(1,3000):
    t=t+dt
    x.append(x[i-1]+dt*v[i-1])
    v.append(v[i-1]-dt*(pot(x[i-1]+dx)-pot(x[i-1]-dx))/(2*dx))
    Ek.append(v[i-1]**2/2)
    V.append(pot(x[i-1]))
    C.append(Ek[i-1]+V[i-1])
```

```python

l1, = plt.plot(V,'b')
l2,l3 = plt.plot(Ek,'r:',C,'g')
plt.legend((l1,l2,l3),('Potencja','Kinetyczna','Cakowita'))
plt.show()
```

```python
l1, = plt.plot(x,'b')
l2, = plt.plot(v,'r')
plt.legend((l1,l2),('Polozenie','Predkosc'))
plt.show()
```

```python
x0=2.8
v0=0
x=x0
v=v0
alf=0
V.clear()
X.clear()
X=[]
V=[]
M=[[1,1],[1,1]]
M[0][1]=-dt/2
M[1][1]=1+dt*alf/2
B=[[1],[1]]
for i in range(1,3000): 
    x=x0
    v=v0 
    for itera in range(0,4):
       
       
        M[1][0]=dt*(pot(x+dx)+pot(x-dx)-2*pot(x))/(2*dx**2)
        B[0]=-(x-x0-dt*v/2-dt*v0/2)
        B[1] = -(v - v0 - dt*(-(pot(x+dx)-pot(x-dx))/(2*dx) - alf*v)/2 - dt*(-(pot(x0+dx)-pot(x0-dx))/(2*dx) - alf*v0)/2)
        #print(B[0])
        #print(B[1],x,v,alf) 
        #print(M)     
        ddx, ddv = scipy.linalg.solve(M,B,overwrite_b=False)
        x=x+ddx
        v=v+ddv
       
        
    #print("end")
    x0=x
    v0=v
    X.append(x)
    V.append(v)
    
        
```

```python
l1, = plt.plot(X,'b')
l2, = plt.plot(V,'r')
plt.legend((l1,l2),('Polozenie','Predkosc'))
plt.show()
```

```python
POT=[]
POT.clear()
Ek.clear()
C.clear()
for i in range(1,3000):
    Ek.append(V[i-1]**2/2)
    POT.append(pot(X[i-1]))
    C.append(Ek[i-1]+POT[i-1])
l1, = plt.plot(POT,'b')
l2,l3 = plt.plot(Ek,'r:',C,'g')
plt.legend((l1,l2,l3),('Potencja','Kinetyczna','Cakowita'))
plt.show()
POT.clear()
Ek.clear()
C.clear()
X.clear()
V.clear()
```
